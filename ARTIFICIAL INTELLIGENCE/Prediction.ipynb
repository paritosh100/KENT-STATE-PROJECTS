```python
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
import os
```

```python
# Load the trained model
# model = load_model('/content/drive/MyDrive/Dataset/flood_earthquake_classifier.h5')
model = load_model('/content/drive/MyDrive/Dataset/flood_earthquake_classifier_finetuned.h5')
```

```python
test_earthquake = '/content/drive/MyDrive/Dataset/Test/Earthquake'

IMG_SIZE = (224, 224)

  # loop through all images in the folder
  for image_name in os.listdir(test_earthquake):
      # Skip like .DS_Store because i use a mac and is a hidden file
      if image_name == '.DS_Store':
          continue
      image_path = os.path.join(test_earthquake, image_name)
      try:
          img = load_img(image_path, target_size=IMG_SIZE)
          img_array = img_to_array(img) / 255.0
          img_array = np.expand_dims(img_array, axis=0)

          # Predict what class the images will be
          prediction = model.predict(img_array)
          print(f"Image: {image_name}")
          print(f"Raw prediction: {prediction[0][0]}")

          if prediction[0][0] > 0.6:
              print(f"Classified as Flood with a confidence of {prediction[0][0] * 100:.2f}%.\n")
          #elif prediction[0][0] < 0.2:
              #print(f"This is neither a Flood nor Earthquake")
          else:
              print(f"Classified as Earthquake with a confidence of {(1 - prediction[0][0]) * 100:.2f}%.\n")#between 0.2 and 0.6

      except Exception as e:
          print(f"Error processing {image_name}: {e}")
```

```python
test_flood = '/content/drive/MyDrive/Dataset/Test/Flood'

IMG_SIZE = (224, 224)

# loop through all images in the folder
for image_name in os.listdir(test_flood):
    if image_name == '.DS_Store':
        continue
    image_path = os.path.join(test_flood, image_name)
    try:
        img = load_img(image_path, target_size=IMG_SIZE)
        img_array = img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # Predict what class the images will be
        prediction = model.predict(img_array)
        print(f"Image: {image_name}")
        print(f"Raw prediction: {prediction[0][0]}")

        if prediction[0][0] > 0.6:
            print(f"Classified as Flood with a confidence of {prediction[0][0] * 100:.2f}%.\n")
        else:
            print(f"Classified as Earthquake with a confidence of {(1 - prediction[0][0]) * 100:.2f}%.\n")

    except Exception as e:
        print(f"Error processing {image_name}: {e}")
```
